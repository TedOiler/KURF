```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
def f(x1, x2):
    return -(x1**2 + x2**2) + 20*np.cos(x1)*np.cos(x2)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ackley(x, y):
    a = 20
    b = 0.2
    c = 2 * np.pi
    sum1 = x**2 + y**2
    sum2 = np.cos(c * x) + np.cos(c * y)
    return a * np.exp(-b * np.sqrt(sum1 / 2)) + np.exp(sum2 / 2) + a + np.exp(1)

# Create a mesh grid
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = ackley(X, Y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.title('Ackley Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def schaffer_n2(x, y):
    num = np.sin(x**2 - y**2)**2 - 0.5
    denom =1 + 0.1*(x**2 + y**2)
    return 20 - 20*num/denom

# Create a mesh grid
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
Z = schaffer_n2(X, Y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.title('schaffer_n2 Function')
plt.show()
```
